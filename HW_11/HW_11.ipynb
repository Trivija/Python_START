```python
from sympy import *
from sympy.plotting import plot
```

```python
x = Symbol('x')
```

```python
f = x**3 - 50*x
```

```python
g = -x**4 + 88*x**2 - 241
```

```python
k = f - g
```

```python
# 1.Координаты точек пересечения.
```

```python
[(el.n().round(2), f.subs(x, el).n().round(2)) for el in solve(k)]
```

```python
# Графики функций в одной системе координат.
```

```python
l = plot(f, g,)
```

```python
# График функции пересечения.
```

```python
m = plot(k)
```
